> **List Comprehension**

```python
lst = []
for x in range(10):
    lst.append(x ** 2)
    
print(lst)
```

```python
# V-1.0 with list comprehension
lst = [x**2 for x in range(10)]

print(lst)
```

```python
# V-2.0 with list comprehension with some conditions
lst = [x**2 for x in range(10) if x%2 == 0]

print(lst)
```

*Nested List Comprehension*

```python
lst1 = [1,2,3,4]
lst2 = ['a','b','c','d']

pair = [(i, j) for i in lst1 for j in lst2]

print(pair)
```

```python
import random

random.seed(0)
random_numbers = [random.randint(1, 20) for _ in range(15)]
print(f"Original list: {random_numbers}")

sorted_numbers = sorted(random_numbers)
print(f"Sorted in ascending order: {sorted_numbers}")

sorted_numbers_desc = sorted(random_numbers, reverse=True)
print(f"Sorted in descending order: {sorted_numbers_desc}")

unique_numbers = list(set(random_numbers))
print(f"List with duplicates removed: {unique_numbers}")
```

```python
students = [
    {'name': 'Alice', 'score': 88},
    {'name': 'Bob', 'score': 72},
    {'name': 'Charlie', 'score': 95},
    {'name': 'David', 'score': 65},
    {'name': 'Eve', 'score': 78}
]

sorted_score = sorted(students, key=lambda x: x['score'], reverse=True)

sorted_score
```

```python
# manage a todo list
to_do_list = ['Buy Groceries', 'Clean the house', 'Laundry', 'Swimming']

## adding the task
to_do_list.append('Blood test')
to_do_list.append('Go for a Run')

# removing a comleted task
to_do_list.remove('Clean the house')

# Checking if the task in to_do_list
if 'Blood test' in to_do_list:
    print('Don\'t forget to get your Blood test done for Full body checkup!!')
    
print('remaing task')
for task in to_do_list:
    print(f'-{task}')
```

```python
# organising students marks
marks = [85, 92 ,78, 90, 88]

# Adding a new grade
marks.append(87)

# calculate the average marks
average_marks = sum(marks) / len(marks)
print(f'Average Marks of the class is {average_marks:.2f}')

# highest and lowest marks of the class
lowest = min(marks)
highest = max(marks)
print(f'Highest marks is {highest}')
print(f'Lowest marks is {lowest}')
```

```python
# Managing an Inventory
inventory = ['apples', 'banana', 'oranges', 'grapes']

# adding a new item
inventory.append('kiwi')

# removing an item that is out of stock now
inventory.remove('banana')

# checking if an item in in stock
item = 'oranges'
if item in inventory:
    print(f'{item } are available in stock')
else:
    print(f'{item} are out of stocks')
    
for items in inventory:
    print(f'-{items}')
```
