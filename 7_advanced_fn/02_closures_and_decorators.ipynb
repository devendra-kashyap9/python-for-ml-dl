### Closures and Decorators

#### Closures:
**A Python closure is a nested function that remembers and retains access to variables from its enclosing (outer) scope, even after the outer function has completely finished executing. It allows you to attach data or context directly to a function without relying on global variables**

---
#### Decorators:
**A decorator in Python is a design pattern that allows you to modify or extend the behavior of a function or method without changing its actual source code**.

> Closures

```python
def fn_welcome(msg):
    def greet():
        print('Welcome here inside the closure function')
        print(msg)
        print("Pay Attention...")
        
    return greet()  # functions can be returned the same way a variable can
```

```python
fn_welcome("Hello Budyy!!!")
```

```python
def fn_welcome(func, lst):
    def greet():
        print('Welcome here inside the closure function')
        print(func(lst))
        print("Pay Attention...")
        
    return greet()
```

```python
fn_welcome(len, [1,2,3,4,5,6])
```

> Decorators

```python
def fn_welcome(func):
    def greet():
        print('Welcome here inside the Python Programming')
        func()
        print("Pay Attention...")
        
    return greet()
```

```python
def fn_introduction():
    print('This is Python Decorators')
    
fn_introduction()
```

```python
fn_welcome(fn_introduction)
```

```python
@fn_welcome   # '@" keyword for decorator"
def fn_introduction():
    print('This is Python Decorators')
```

```python
## Decorators with argumrnts
def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator
```

```python
@repeat(3)
def say_hello():
    print('Hello!!') 
```

```python
say_hello()
```

```python
def log_decorator(func):
    def wrapper(*args, **kwargs):
        print(f"Running {func.__name__} with arguments {args}")
        result = func(*args, **kwargs)
        print(f"{func.__name__} finished executing.")
        return result
    return wrapper

@log_decorator
def add_numbers(a, b):
    return a + b

print(add_numbers(5, 10))
```
