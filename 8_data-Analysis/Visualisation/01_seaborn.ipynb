```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# basic plotting with seaborn
tips = sns.load_dataset('tips')
tips.head()
```

```python
# scatterplot
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter plot of Total_bill vs Tips')
plt.show()
```

```python
# line plot
sns.lineplot(x='size', y='total_bill', data=tips)
plt.title('Line plot of Total_bill by size')
plt.show()
```

```python
# categorical plot
# bar plot
sns.barplot(x='day', y='total_bill', data=tips)
plt.title('Bar plot of Total_bill by day')
plt.show()
```

```python
# box plot
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Box plot of Total_bill by day')
plt.show()
```

```python
# violin plot
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('violin plot of Total_bill by day')
plt.show()
```

```python
# histogram
sns.histplot(tips['total_bill'], 
             bins=10,
             kde=True)
plt.title('Hist plot of Total_bill')
plt.show()
```

```python
# kde plot
sns.kdeplot(tips['total_bill'], 
            fill=True);
```

```python
# pair plot
# relationship b/w every feature
sns.pairplot(tips);
```

```python
# heatmap
corr = tips[['total_bill', 'tip', 'size']].corr()
corr
```

```python
sns.heatmap(corr,
            annot=True,
            cmap='coolwarm');
```

```python
# sales data
import pandas as pd
df = pd.read_csv('salesdata.csv')
df.head()
```

```python
produc_total_sales = df.groupby('Product')['Sales'].sum()
produc_total_sales
```

```python
# plot total sales by product
plt.figure(figsize=(10,6))
sns.barplot(x='Product', y='Sales', data=df,
            estimator=sum)
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.title('Total Sales By Product')
plt.show()
```

```python
# plot total sales by region
plt.figure(figsize=(10,6))
sns.barplot(x='Region', y='Sales', data=df,
            estimator=sum)
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.title('Total Sales By Region')
plt.show()
```
