### 🧬 Object-Oriented Programming (OOP) in Python: Classes & Objects

Object-Oriented Programming (OOP) is a programming paradigm that organizes software design around **data**, or **objects**, rather than functions and logic. It treats a program as a collection of interacting objects that package both data and behavior together.

---

#### 1. Core Concepts: The Blueprint vs. The Instance

To understand OOP, you must master the relationship between a **Class** and an **Object**. 

*   **Class (The Blueprint):** A user-defined prototype or template. It defines the structure, variables (attributes), and behaviors (methods) that all objects of this type will have. It occupies no memory space on its own.
*   **Object (The Instance):** A concrete, real-world entity created from the class blueprint. It occupies actual physical memory space and holds real data unique to that specific instance.

### 🚗 The Analogy
Think of a **Class** as the architectural blueprint of a car. It specifies that a car must have 4 wheels, a specific color, and the ability to accelerate. An **Object** is the actual physical car (e.g., a Red 2026 Tesla Model 3) built from that blueprint that you can physically drive.

---

#### 2. Anatomy of a Python Class

When creating a class in Python, we use specific components to handle state initialization and functional behaviors.

| Component | What it represents | Python Syntax Reference |
| :--- | :--- | :--- |
| **Class Attribute** | Data shared by **all** instances of the class. | Declared directly inside the class body. |
| **Instance Attribute** | Data unique to **each specific instance**. | Declared inside the `__init__` constructor method. |
| **Constructor (`__init__`)** | Special initializer method called automatically when an object is created. | `def __init__(self, ...):` |
| **`self` Parameter** | A reference to the **current instance** being worked on. | Must be the first parameter of any instance method. |
| **Method** | A function defined inside a class that represents object behaviors. | Regular function syntax indented within the class. |



```python
# A class is a blueprint for creating objects, Attributes and methods.
class Car(object):
    pass

audi = Car()
bmw = Car()

print(type(audi))
print(type(bmw))
print(type(Car))
```

```python
audi.windows = 4
print(audi.windows)
```

```python
# instance variable and methods
class Dog:
    # constructoe/initializer
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def __str__(self):
        return f'name: {str(self.name)} and age: {str(self.age)}'
    

# create objects of Dog
dog1 = Dog('Buddy', 3)
print(dog1)
print(dog1.name)
print(dog1.age)
```

```python
dog2 = Dog('Lucy', 4)
print(dog2)
print(dog2.name)
print(dog2.age)
```

```python
# class with instance methods
class Dog:
    # constructoe/initializer
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def __str__(self):
        return f'name: {str(self.name)} and age: {str(self.age)}'
    
    def bark(self):
        print(f'{self.name} says Woof...')
        
dog1 = Dog('Buddy', 3)
dog1.bark()
dog2 = Dog('Lucy', 5)
dog2.bark()
```

```python
# Modelling a bank account

# define a class for bank
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance
        
    def deposit(self, amount):
        self.balance += amount
        print(f"{amount} deposited, new balance is {self.balance}")

    def withdraw(self, amount):
        if self.balance < amount:
            print('Insufficient Balance...')
        else:
            self.balance -= amount
            print(f'{amount} is withdrawn, remaining balance is {self.balance}')  
            
    def get_balance(self):
        return self.balance
    
# create an account
account = BankAccount('Alex', 5000)
print(account.get_balance())  
```

```python
# call instance methods
account.deposit(100)
```

```python
account.withdraw(50) 
```

```python
print(account.get_balance()) 
```
