### 🐍 Python Exception Handling: Core Theories

In programming, things don't always go as planned. Exception handling is the safety net that prevents your Python scripts from crashing when they encounter unexpected conditions at runtime.

---

#### 1. Syntax Errors vs. Exceptions

Before writing error-handling code, you must distinguish between the two types of errors Python throws.

##### A. Syntax Errors (Parsing Errors)
These happen when Python reads your code and finds that you broke the structural rules of the language. The code **fails before it even starts running**.
*   *Example:* Missing a colon `:` at the end of a `for` loop or having unmatched parentheses.

##### B. Exceptions (Runtime Errors)
These occur when your code is structurally perfect, but an unexpected state or invalid operation happens **while the program is executing**.
*   *Example:* Trying to read a file that has been deleted, or dividing a number by zero.

---

#### 2. The Core Exception Architecture

Python treats every exception as an **object**. All exceptions belong to a built-in hierarchy, branching down from the master `BaseException` class.

##### Common Built-in Exceptions to Know:
*   `ZeroDivisionError`: Raised when the second argument of a division or modulo operation is zero.
*   `ValueError`: Raised when a function receives an argument of the correct type but an inappropriate value (e.g., trying to turn `"apple"` into an integer via `int()`).
*   `TypeError`: Raised when an operation or function is applied to an object of an inappropriate type (e.g., adding a string to an integer: `'3' + 5`).
*   `FileNotFoundError`: Raised when a file or directory is requested but doesn't exist.
*   `IndexError`: Raised when a sequence subscript (like a list index) is out of range.

---

#### 3. The Anatomy of the `try-except-else-finally` Block

Python handles exceptions using a 4-part block structure. While only `try` and `except` are required, utilizing `else` and `finally` provides robust control over your application's flow.

| Block Layer | Operational Goal | Execution Condition |
| :--- | :--- | :--- |
| **`try`** | Monitors risky code | **Always** executes first. |
| **`except`** | Catches and mitigates errors | Executes **only if** an exception occurs inside the `try` block. |
| **`else`** | Executes "success-only" code | Executes **only if** the `try` block ran with zero errors. |
| **`finally`** | Performs necessary clean-up tasks | **Always** executes, regardless of whether an error occurred or was caught. |


```python
try:
    a = b
except:
    print('The variable has not been assigned yet')
```

```python
try:
    a = b
except NameError as nm:
    print(nm)
```

```python
try:
    result = 1/2
    a = b
except ZeroDivisionError as zde:
    print(zde)
except Exception as ex:  # Parent exception class Exception
    print(ex)
```

```python
try:
    num = int(input('Enter a number: '))
    result = 10 / num
except ValueError as ne:
    print(ne)
except ZeroDivisionError as zde:
    print(zde)
except Exception as ex:
    print(ex)
```

```python
# try except else
try:
    num = int(input('Enter a number: '))
    result = 10 / num
except ValueError as ne:
    print(ne)
except ZeroDivisionError as zde:
    print(zde)
except Exception as ex:
    print(ex)
else:
    print(result)
```

```python
# try except else finally
try:
    num = int(input('Enter a number: '))
    result = 10 / num
except ValueError as ne:
    print(ne)
except ZeroDivisionError as zde:
    print(zde)
except Exception as ex:
    print(ex)
else:
    print(result)
finally:
    print('The code execution is over now...')
```
