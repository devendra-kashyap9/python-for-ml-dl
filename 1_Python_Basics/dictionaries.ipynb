### Dictionaries

Dictionaries are the unorderd collections of `key:value` pairs. Key must be unique and immutable(string, numbers, tuples).

```python
# creating dictionary
my_dict = {}
print(type(my_dict))
```

```python
empty_dict = dict()
print(type(empty_dict))
```

```python
student = {"name": "Alex",
            "Age": 23,
            'Grade': "A"}

print(student)
```

```python
# is key is not unique
student = {"name": "Alex",
            "Age": 23,
            'name': "A"}

print(student)
```

```python
# accessing dict elements 
student = {"name": "Alex",
            "Age": 23,
            'Grade': "A"}

print(student['Grade'])
print(student["name"])
```

```python
# get method
print(student.get('name'))
print(student.get('Grade'))

# if key os not present
print(student.get('Address'))

# we can give default value instead of getting NOne
print(student.get('Address', 'Not Available'))
```

```python
# Modifying dict elemants 
print(student)
```

```python
student['Age'] = 25  # updation
student['Address'] = 'India'  # insertion
print(student)
```

```python
# delete the key
del student['Grade']
print(student)
```

### dict Methods

```python
# Getting all the keys
keys = student.keys()
print(keys)

# getting all the values
values = student.values()
print(values)

# getiing all the key val pairs
items = student.items()
print(items)
```

```python
# shallow copy
student_copy = student
print(student)
print(student_copy)
```

```python
student['name'] = 'Bob'
print(student)
print(student_copy)
```

Means our last initialisation is reflecting the change in both lets tweak and work upon.

```python
student_copy1 = student.copy()
print(student)
print(student_copy)
```

```python
student['name'] = 'Charles'
print(student)
print(student_copy1)
```

```python
# Iterating over dict
for key, val in student.items():
    print(key, val)
```

```python
for key in student.keys():
    print(student[key])
```

```python
for val in student.values():
    print(val)
```

```python
# Nested dict
students = {
    'Student1': {'name':"Dia", 'age':25},
    'Student2': {'name':'Eva', 'age': 24}
}

print(students)
```

```python
# accessing nested dict
print(students['Student1']['name'])
print(students['Student2']['name'])
print(students['Student1']['age'])
```

```python
# Iterating over nested dict
for student_id, student_info in students.items():
    print(student_id, student_info)
    for key, val in student_info.items():
        print(key, val)
```

### Dictionary comprehension

```python
squares = {x:x**2 for x in range(5)}
print(squares)
```

```python
# conditional dict comprehension
even_square = {x:x**2 for x in range(11) if (x%2 == 0)}
print(even_square)
```

### Task

```python
# use dictionary to count the frequency in a list
lst = [1,2,3,4,1,2,3,1,3,4]
frequency = {}

for number in lst:
    if number in frequency:
        frequency[number] += 1
    
    else:
        frequency[number] = 1

print(frequency)
```

```python
# alternative
from collections import Counter

frequency = Counter(lst) 
print(frequency)
```

```python
# merging the dict
dict1 = {'a':1, 'b':2}
dict2 = {'c':3, 'd':4}

merged_dict = {**dict1, **dict2}
print(merged_dict)
```
