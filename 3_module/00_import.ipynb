### Importing Modules in Python: Modules and Packages

In Python, modules and packages help origanize and reuse code. Here's comprehesive guide on how to import them.

```python
import math
math.sqrt(60)
```

```python
from math import sqrt, pi
print(sqrt(25))
print(sqrt(16))
print(pi)
```
