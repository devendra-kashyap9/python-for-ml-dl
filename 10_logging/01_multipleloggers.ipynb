### Logging with multiple loggers

```python
import logging

# create a logger for module 1
logger1 = logging.getLogger('module1')
logger1.setLevel(logging.DEBUG)

# create a logger for module 2
logger2 = logging.getLogger('module2')
logger2.setLevel(logging.WARNING)

# Configure logging to write to a file with a specific format
logging.basicConfig(
    level=logging.DEBUG, # Capture EVERYTHING from DEBUG level up
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)
```

```python
# log message with different severity level
logger1.debug('THis is debug msg foe Module1')
logger2.warning('This is warning message for Module2')
logger2.error('This is an error message')
```
