### Tuple Operation

### Immutable nature of Tuples

Tuples are immutable means the elements of tuples cannot be changed once `assigned`.

```python
lst = [1,2,3,4]
print(lst)

lst[0] = 5
print(lst)
print(type(lst))

# tuple
tupe = (1,2,3,4)
print(tupe)
print(type(tupe))

tupe[0] = 5
print(tupe)
```

### Tuple methods

```python
# count method
numbers = (1,2,3,4,5,1)
print(numbers.count(1))  # total occurence/frequency of an element
print(numbers.index(4))  # at which index the passed element is present
```

```python
# Packing and unpacking of tuples
packed_tuple = 1, "Hello World", 4.5
print(packed_tuple)
```

```python
# Unpacking a tuple
a, b, c = packed_tuple
print(a)
print(b)
print(c) 
```

```python
# unpacking with *
numbers = (1,2,3,4,5,6)
start, mid, end = numbers
print(start)
print(mid)
print(end)
```

```python
start, *mid, end = numbers
print(start)
print(mid)
print(end)
```

```python
# Nested tuples
tupe = ((1,2,3),
        (4,5,6),
        (7,8,9))

print(tupe[0])
print(tupe[0][2])
print(tupe[1][:2])
```

```python
# Iterating with tutples
for sub_tuples in tupe:
    for val in sub_tuples:
        print(val, end=' ')
    print('\n')
```
