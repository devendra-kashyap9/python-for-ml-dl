### File Operations: Read and Write

File Handling is a crucial part of programming language. Python provides built in functions and methods to `read from` and `write to` files, both text and binary.

```python
## Read an entire file
with open('examples.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
# as we do not have this file it throws an error lets create the file and rerun
## Read an entire file
with open('examples.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
# read line by line
with open('examples.txt', 'r') as file:
    for line in file:
        print(line.strip())  # .strip() will remove extra new line character
```

```python
# writing into a file
with open('examples.txt', 'w') as file:
    file.write('Hello World: \n')
    file.write('this is a line and overwrite the file')
```

```python
# writting in such a way that it won't overwrite
# now open a file in append mode
with open('examples.txt', 'a') as file:
    file.write('This is appended new line\n')
    file.write('Enjoy the day')
```

```python
# Writing list of lines
lines = ['first line \n', 
         'second line \n',
         'last line \n']
with open('examples.txt', 'a') as file:
    file .writelines(lines)
```

### Binary Files

```python
# writing to a binary file
data = b'\x00\x01\x02\x03\x04'
with open('examples.bin', 'wb') as file:
    file.write(data)
```

```python
with open('examples.bin', 'rb') as file:
    content = file.read()
    print(content)
```

### 📄Read the content from source text file and write it into the destination text file.

```python
with open('examples.txt', 'r') as source_file:
    content = source_file.read()

with open('destination.txt', 'w') as destination_file:
    destination_file.write(content)
```

```python
# Writing and then reading the file
'''The w+ mode in Pythib is used to open a file for both read and write. If the file does not exist in will be
created, else its content get truncated i.e overwritten'''

with open('examples.txt', 'w+') as file:
    file.write('Hello World\n')
    file.write('this is new line')
    
    # move the file cursor to begining
    file.seek(0)
    
    # read the content of file
    content = file.read()
    print(content)
```

```python
# create a new directory
import os 

new_directory = 'package'
os.mkdir(new_directory)
print(f'Directory "{new_directory}" created...')
```

```python
## Listing files and dirs
items = os.listdir('.')  # . is current location
print(items)
```

```python
# joining path
# relative path
dir_name = 'folder'
file_name = 'file.txt'
full_path = os.path.join(dir_name, file_name)
print(full_path)
```

```python
# joining path
# full path
dir_name = 'folder'
file_name = 'file.txt'
full_path = os.path.join(os.getcwd(), dir_name, file_name)
print(full_path)
```

```python
# checking the path exists
path = 'example.txt'
if os.path.exists(path):
    print(f'The path {path} exists')
else:
    print(f'The path {path} doesn\'t exists')
```

```python
path = 'examples.txt'

if os.path.isfile(path):
    print(f'The path {path} is a file')
elif os.path.isdir(path):
    print(f'The path {path} is a directory')
else:
    print(f'The path {path} is a neither a file nor a directory')
```

```python
# Getting the absolute path
relative_path = 'examples.txt'
absolute_path = os.path.abspath(relative_path)
print(absolute_path)
```
