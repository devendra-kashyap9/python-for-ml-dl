### Standard library overview

Python's standard library is a vast collection of module and packages that comes bundled with python, providing a wide range of functionalitied of thr box. Here's an overview of some of the most commonly used modules and packages in the Python Standard Library.

```python
import array
arr = array.array('i', [1,2,3,4]) # 'i' is typecode for an integer
print(arr)
```

```python
import math
math.sqrt(16)
```

```python
# random

import random
print(random.randint(1,11))
print(random.choice(['apple', 'banana', 'cherry']))
print(random.choices(['apple', 'banana', 'cherry'], 
                     k=2,       # number of elements we want from the list
                     weights=[50, 40, 10])) # probability of individual element to occur
```

```python
# file and directory access
# os library

import os
print(os.getcwd())
```

```python
# high level operations on files and collections of files
import shutil
shutil.copyfile('source.txt', 'dest.txt')
```

```python
# Data Serialization
import json
data = {'name': 'Alex', 'age': 26}

json_str = json.dumps(data)
print(json_str)
print(type(json_str))

# convert back to original form
parsed_data = json.loads(json_str)
print(parsed_data)
print(type(parsed_data))
```

```python
# csv
import csv

with open('examples.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['name', 'age'])
    writer.writerow(['Dia', 29])
    
with open('examples.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
        
file.close()
```

```python
# datetime 
from datetime import datetime, timedelta

now = datetime.now()
print(now)  # current time

yesterday = now - timedelta(days=1) # time one day back
print(yesterday)
```

```python
# import time
import time
print(time.time())
time.sleep(2)
print(time.time())
```

```python
# regular expression
import re

pattern = r'\d+'
text = 'There are 123 appple'
match = re.search(pattern, text)
print(match.group())
```
