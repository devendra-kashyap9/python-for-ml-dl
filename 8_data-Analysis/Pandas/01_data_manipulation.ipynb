```python
# reading the data from csv file
df = pd.read_csv('sales_data.csv', encoding_errors='ignore')

df.head()
```

```python
## Handling missing values
df.isna().sum()
```

```python
df.isna().any()
```

```python
# filling missing values with default values
df_filled = df.fillna(0)
df_filled.isna().sum()
```

```python
# filling missing val with the mean or median of that column, if data is "Numerical"
# for categorical data 'mode' could be the better option 
# we also have one more option 'most frequesnt' but we dp that with Pipeline and ColumnTransformer
df['City_filled'] = df.CITY.fillna(df.CITY.mode())
df.CITY.mode()
```

```python
df.dtypes
```

```python
# renaming columns
df = df.rename(columns={'SALES':'Total_Sales'})  # we can pass multiple columns
df.head()
```

```python
# change data types
df['new_values'] = df.Total_Sales.astype(int)
df.head()
```

```python
df['new_quantity'] = df.QUANTITYORDERED.apply(lambda x: x*2)
df[['QUANTITYORDERED', 'new_quantity']].head()
```

```python
# Data Agrregating and grouping
df.head()
```

```python
pd.unique(df['CITY'])
```

```python
# mean of totalsales by city
mean_by_city = df.groupby('CITY')['Total_Sales'].mean()
mean_by_city.head(), len(mean_by_city)
```

```python
grouped_sum = df.groupby(['CITY', 'STATUS'])['Total_Sales'].sum()
grouped_sum
```

```python
# applying multiple aggregate function
agg_groupby = df.groupby('CITY')['Total_Sales'].agg(['mean', 'sum', 'count'])
agg_groupby.head()
```

```python
## Merging and joining dataframes
df1 = pd.DataFrame(
    {'Keys': ['A', 'B', 'C'],
     'Value1': [1, 2, 3]}
)

df2 = pd.DataFrame(
    {'Keys': ['A', 'B', 'D'],
     'Value2': [4, 4, 6]}
)
```

```python
df1
```

```python
df2
```

```python
df1.merge(df2)
```

```python
# merging the dataframe on key column
pd.merge(df1, df2,
         on='Keys',
         how='inner')
```

```python
pd.merge(df1, df2,
         on='Keys',
         how='outer')
```

```python
pd.merge(df1, df2,
         on='Keys',
         how='left')
```

```python
pd.merge(df1, df2,
         on='Keys',
         how='right')
```

```python
# concatenating
# Pass df1 and df2 inside a list
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], ignore_index=True)
```
