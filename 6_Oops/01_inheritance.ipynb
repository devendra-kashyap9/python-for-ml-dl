### 🌿 Inheritance Theory in Python

**Inheritance** is one of the four foundational pillars of Object-Oriented Programming (OOP). It allows a new class to inherit the attributes and methods of an existing class. 

*   **Parent Class** (also called *Superclass* or *Base Class*): The existing class whose properties are being inherited.
*   **Child Class** (also called *Subclass* or *Derived Class*): The new class that inherits features from the parent class.

---

#### 1. Core Mechanics: The "Is-A" Relationship

Inheritance models a clean **"Is-A"** relationship. For example:
*   A `Car` **is a** `Vehicle`.
*   A `Manager` **is an** `Employee`.
*   A `SavingsAccount` **is a** `BankAccount`.

##### Why Use Inheritance?
1.  **Code Reusability:** You write common logic once in the parent class and reuse it across multiple child classes without copying code.
2.  **Extensibility:** You can add unique attributes or behaviors to a child class without touching or breaking the parent class code.
3.  **Maintainability:** If a bug is found in a shared behavior, fixing it in the parent class automatically fixes it across all subclasses.

---

#### 2. Basic Syntax and the `super()` Function

In Python, you pass the parent class name inside parentheses right next to the child class name during definition.

To invoke the parent class's constructor (`__init__`) or its methods from inside the child class, we use the built-in **`super()`** function. This ensures that the parent state is fully initialized before the child layer adds its unique modifications.

```python
# Parent class
class Car:
    def __init__(self, windows, doors, enginetype):
        self.windows = windows
        self.doors = doors
        self.enginetype = enginetype
        
    def drive(self):
        print(f'The person is driving {self.enginetype} car') 
```

```python
car1 = Car(4, 5, "Diesel")
car1.drive()
```

```python
# child/derived class
# Single Inheritance - when a child class inherits from a single base class
class BMW(Car):
    def __init__(self, windows, doors, enginetype, is_selfdriving):
        super().__init__(windows, doors, enginetype)
        self.is_selfdriving = is_selfdriving
        
    def self_driving(self):
        print(f'BMW supports self driving: {self.is_selfdriving}')
```

```python
bmw = BMW(4, 5, 'Hybrid', False)
bmw.drive()
bmw.self_driving()
```

```python
# Multiople Inheritance - when a child class inherits from multiple base classes.
# base class 1
class Animal:
    def __init__(self, name):
        self.name = name
        
    def speak(self):
        print('Subclasses must implements this method')
        
# base class 2
class Pet:
    def __init__(self, owner):
        self.owner = owner
        
# Derived class
class Dog(Animal, Pet):
    def __init__(self, name, owner):
        Animal.__init__(self, name)  # as we have 2 base class it is wise to call the init with class name rather
        Pet.__init__(self, owner)    # just calling with super
        
    def speak(self):
        return f'{self.name} says Woof...'
        
# create an object
dog = Dog('Buddy', 'Dia')
print(f'{dog.name = }')
print(f'{dog.owner = }')
print(dog.speak())        
```
