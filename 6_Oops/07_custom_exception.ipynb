### Custom Exception `(Raise and Throw)` an exception

```python
# defining some custom Exceptions

class Error(Exception):
    pass

class dobException(Error):
    pass
```

```python
year = int(input('Enter dob'))
age = 2026 - year

try:
    if age <= 30 and age >= 20:
        print("The age is valid and you are eligible for the exam")
    else:
        raise dobException
    
except dobException:
    print('Sorry, you are ineligible now..')
    
```
