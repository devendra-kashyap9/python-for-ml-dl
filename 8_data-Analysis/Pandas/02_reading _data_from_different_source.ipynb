### reading data from different sources

```python
import json
import pandas as pd 
from io import StringIO

data = {
    "id": 101,
    "product_name": "Wireless Mouse",
    "category": "Electronics",
    "price": 29.99,
    "in_stock": True,
    "tags": ["computer", "accessory", "wireless"]
  }  
# Convert Python dict/list to a JSON string
json_string = json.dumps(data)

# Force pandas to read it correctly as a single series/object
pd.read_json(StringIO(json_string), typ='series')
```

```python
data = """{
    "employee_name": "James",
    "email": "james@email.com",
    "title": [{"title1": "team_lead", "title2": "Senior_developer"}]
}"""

df = pd.read_json(StringIO(data))
df
```

```python
df.to_json()
```

```python
df.to_json(orient='index') # default
```

```python
df.to_json(orient='records')
```

```python
import pandas as pd

url = "https://en.wikipedia.org/wiki/Mobile_country_code"

# Fake a real browser user-agent string
custom_headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

# Pass the headers into storage_options
df_list = pd.read_html(url, 
                       match="Country", 
                       header=0, 
                       storage_options=custom_headers)

# Access the first matched table
df_list[0]
```
