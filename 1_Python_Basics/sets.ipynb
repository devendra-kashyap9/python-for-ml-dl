### Sets

* Sets are built-in data type in Pythin used to store unique items.
* They are unordered collections.

They are useful for `membership tests, eliminating duplicate enteries and for performing mathematical set ops`, like *union, intersection, difference and symmetric difference*.

```python
# creating a set
first_set = {1,2,3,4,5}
print(first_set)
print(type(first_set))
```

```python
empty_set = set()
print(empty_set)
```

```python
new_set = set([1,2,3,4])
print(new_set)
```

```python
# sets are unique means doesn't allow duplicacy
new_set = set([1,2,3,4,3,2,4,1,1,2])
print(new_set)
```

### Basic set operation

```python
## adding and removing the element in set
my_set = {1,2,4,3,5,8,7,4}

my_set.add(9)
print(my_set)
my_set.add(9)
print(my_set)
```

```python
# remove an element
my_set.remove(3)
print(my_set)
```

```python
# removing an element which is no present
my_set.remove(10)
```

```python
# we use "discard" method
# It will remove the element if present leave otherwise
my_set.discard(10)
print(my_set)
```

```python
# pop
poped_element = my_set.pop()
print(poped_element)
print(my_set)
```

```python
# removing all the elements from a set
my_set.clear()
print(my_set)
```

### Set membership test

```python
my_set = {1,2,3,4,5}
print(3 in my_set)
print(10 in my_set)
```

### Mathematical operation

```python
set1 = {1,2,3,4,6}
set2 = {2,5,7,9,6}
```

```python
# union
union_set = set1.union(set2)
print(union_set)
```

```python
# Intersection 
intersection_set = set1.intersection(set2)
print(intersection_set)
```

```python
# intersection_update
print(set1)
set1.intersection_update(set2)  # intersection_update gives the common elements and also update 
                                # the set from which it is called upon
print(set1)
```

```python
set1 = {1,2,3,4,6}
set2 = {2,5,7,9,6}

# difference
print(set1.difference(set2))   # gives non common element of collee set
```

```python
set1 - set2  
```

```python
# Symmetric difference
set1.symmetric_difference(set2)  # remove the common element and combine
```

### Sets Methods

```python
set1 = {1,2,3,4,5,6}
set2 = {4,5,6}
```

```python
# is subset
print(set1.issubset(set2))
```

```python
# is superset
print(set1.issuperset(set2))
```

### Counting unique words in set

```python
text = "In this notebookd we are discussing about sets and its operations."
words = text.split()
print(type(words))

# convert list of words to set to get unique words
unique_words = list(set(words))
print(unique_words)
```
