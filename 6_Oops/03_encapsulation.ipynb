### Encapsulation in Python: Theory & Core Concepts

**Encapsulation** is one of the four fundamental pillars of Object-Oriented Programming (OOP) (alongside Inheritance, Polymorphism, and Abstraction). It describes the bundle of data (attributes) and the methods that operate on that data into a single unit—a **Class**. 

Furthermore, encapsulation restricts direct access to some of an object's components, which is crucial for preventing accidental data modification and maintaining a clean system architecture.



---

#### 1. Why Do We Need Encapsulation?

1. **Data Hiding / Security:** Internal object states are protected from unauthorized direct modifications by external code.
2. **Flexibility and Maintenance:** You can change the internal implementation of a class without breaking the code that relies on it. 
3. **Control:** You can enforce validation rules (e.g., ensuring an `age` attribute can never be set to a negative number).

---

#### 2. Access Modifiers in Python

Unlike languages like Java or C++ which have explicit keywords (`public`, `private`, `protected`), Python manages accessibility using a naming convention with **underscores**.

| Type | Naming Convention | Access Level | Description |
| :--- | :--- | :--- | :--- |
| **Public** | `name` | Anywhere | Accessible from inside or outside the class. |
| **Protected** | `_name` (Single `_`) | Internal & Subclasses | A warning to other developers: *“Treat this as internal; do not modify outside the class/subclass.”* |
| **Private** | `__name` (Double `__`) | Internal Only | Strictly hidden from outside access. Python alters its name via **Name Mangling**. |

---

#### 3. Implementing Encapsulation (Code Structure)

To see how this works inside a Jupyter environment, consider a `BankAccount` class where the balance must be protected from direct, external tampering.

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner            # Public attribute
        self._account_type = "Savings" # Protected attribute
        self.__balance = balance      # Private attribute (Encapsulated)

    # Getter Method (To safely read private data)
    def get_balance(self):
        return self.__balance

    # Setter Method (To safely update private data with validation)
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited {amount}. New Balance: {self.__balance}")
        else:
            print("Invalid deposit amount!")

````python
# Encapsulation with Getter and Setter methods
# access specifier in other languages: private, protected, public

class Person:
    def __init__(self, name, age):
        self.name = name  # public var
        self.age = age  # public var, i.e. it can be accessed evn outside the class scope
        
person = Person('Eva', 27)
print(person.name)
print(person.age)
````

````python
class Person:
    def __init__(self, name, age, sex):
        self._name = name  # protected var
        self.__age = age  # private var, i.e. it can't be accessed  outside the class scope
        self.sex = sex  # public var
        
person = Person('Eva', 27, 'Female')
````

````python
print(person.sex)
print(person._name)
print(person.__age) # private var not accessible outside the class
````

````python
# encapsulatio with getter and setter
class Person:
    def __init__(self, name, age):
        self.__name = name  # Private var
        self.__age = age  # Private var
        
    # getter maethod for name
    def get_name(self):
        return self.__name
    
    # setter method for name
    def set_name(self, name):
        self.__name = name
    
    # getter method for age
    def get_age(self):
        return self.__age
    
    # setter method for age
    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print('Age cannot be negative')
            
person = Person('Diva', 27)

## access and modify the private var with getter and setter method
print(person.get_name())
print(person.get_age())
        
person.set_age(28)
print(person.get_age())     

person.set_age(-10)
````
