### SQL and SQLite

`SQL` (Structured Query Language) is the standard programming language used to communicate with, manage, and query relational databases. It allows you to store, update, remove, and retrieve data organized in tabular rows and columns.

`SQLite` is a C-language library that implements a small, fast, self-contained, and highly reliable SQL database engine. It is serverless and embeds directly into applications, making it the most widely deployed database globally.

**The primary difference is that SQL is a programming language, whereas SQLite is a database engine software that reads and executes that language.Think of SQL as a language (like English) and SQLite as a specific book written in that language**.

```python
## python has built-in library named `sqlite3`
import sqlite3
```

```python
# connect to sqlite database
# lets connect to database by creating a databse
connection = sqlite3.connect('example.db')
connection
```

```python
# create a cursor object
# it will help in iterating over elements/entries
cursor=connection.cursor()
```

```python
# create a table
cursor.execute('''
               CREATE TABLE IF NOT EXISTS employees(
                   id INTEGER PRIMARY KEY,
                   name Text NOT NULL,
                   age INTEGER,
                   department TEXT)
''')

# commit the changes
connection.commit()
```

```python
cursor.execute('''
               SELECT * FROM employees;
''')
```

```python
# Insert the data/entries into table
cursor.execute('''
               INSERT INTO employees(name, age, department)
               VALUES
               ('ALEX', 27, 'Philosophy'),
               ('Bob', 26, 'Yoga'),
               ('Charlie', 26, 'Psychology'),
               ('Dia', 27, 'STEM'),
               ('Eva', 27, 'STEM')              
''')

connection.commit()
```

```python
# fetch the data from tables
cursor.execute('''
               SELECT * FROM employees;
''')
rows = cursor.fetchall()

# print the queried data
for row in rows:
    print(row)
```

```python
# update the data in table
cursor.execute('''
               UPDATE employees
               SET age = 26
               WHERE name LIKE 'EVA';
''')
connection.commit()
```

```python
cursor.execute('''
               SELECT * FROM employees;
''')
rows = cursor.fetchall()

# print the queried data
for row in rows:
    print(row)
```

```python
# delete the data from tables
cursor.execute('''
               DELETE FROM employees
               WHERE name LIKE 'Bob';
''')
connection.commit()
```

```python
cursor.execute('''
               SELECT * FROM employees;
''')
rows = cursor.fetchall()

# print the queried data
for row in rows:
    print(row)
```

```python
# working with sales dataset
connection = sqlite3.connect('sales_data.db')
cursor = connection.cursor()

# Create a tbale for sales data
cursor.execute('''
               CREATE TABLE IF NOT EXISTS sales_data(
                   id INTEGER PRIMARY KEY,
                   date TEXT NOT NULL,
                   product TEXT NOT NULL,
                   sales INTEGER,
                   region TEXT)
''')

# insert some data
sales_data = [
    ('2023-01-01', 'Product1', 100, 'North'),
    ('2023-01-02', 'Product2', 200, 'South'),
    ('2023-01-03', 'Product1', 150, 'East'),
    ('2023-01-04', 'Product3', 250, 'West'),
    ('2023-01-05', 'Product2', 300, 'North')
]


cursor.executemany('''
               INSERT INTO sales_data(date, product, sales, region)
               VALUES(?,?,?,?)                            
''', sales_data)

connection.commit()
```

```python
# get all the records
cursor.execute('''
               SELECT * FROM sales_data;
''')
rows = cursor.fetchall()

# print the queried data
for row in rows:
    print(row)
```
