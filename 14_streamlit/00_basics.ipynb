### Introduction to Streamlit

**`Streamlit` is an open-source Python library used to convert data scripts into interactive web applications in minutes. It requires zero frontend experience, meaning you do not need to know HTML, CSS, or JavaScript to create polished user interfaces**.

Data scientists, machine learning engineers, and developers use it to quickly build interactive dashboards, prototype applications, and share data-driven insights.

#### 💡 Core Concept: How It Works

##### Streamlit uses a unique and simple mental model:

* **Pure Python**: You write standard Python code from top to bottom.
* **The Rerun Execution**: Every time a user interacts with a widget (like moving a slider or clicking a button), Streamlit reruns the entire Python script from the beginning.
* **No Event Handlers**: You don't need to write complex callback functions to handle events. If a user changes a selection box, the variable storing that selection updates automatically on the next rerun.
