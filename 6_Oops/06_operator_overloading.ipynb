```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)
    
    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)
    
    def __mul__(self, other):
        return Vector(self.x * other.x, self.y * other.y)
    
    def __eq__(self, other):
        return Vector(self.x == other.x and self.y == other.y)
    
    def __repr__(self):
        return f"Vector({self.x}, {self.y})"
    
# Create objects of vector class
v1 = Vector(2, 3)
v2 = Vector(4, 5)
v3 = Vector(6, 7)

print(v1 + v2 + v3)
print(v3 - v2)
print(v1 * v2)
```

```python
class ComplexNumber:
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag

    def __repr__(self):
        """Formal string representation for Jupyter display."""
        return f"ComplexNumber({self.real}, {self.imag})"

    def __str__(self):
        """User-friendly string representation."""
        sign = "+" if self.imag >= 0 else "-"
        return f"{self.real} {sign} {abs(self.imag)}i"

    def __add__(self, other):
        """Overloads the '+' operator."""
        if not isinstance(other, ComplexNumber):
            return NotImplemented
        return ComplexNumber(self.real + other.real, self.imag + other.imag)

    def __sub__(self, other):
        """Overloads the '-' operator."""
        if not isinstance(other, ComplexNumber):
            return NotImplemented
        return ComplexNumber(self.real - other.real, self.imag - other.imag)

    def __mul__(self, other):
        """
        Overloads the '*' operator.
        Formula: (a+bi)(c+di) = (ac - bd) + (ad + bc)i
        """
        if not isinstance(other, ComplexNumber):
            return NotImplemented
        real_part = (self.real * other.real) - (self.imag * other.imag)
        imag_part = (self.real * other.imag) + (self.imag * other.real)
        return ComplexNumber(real_part, imag_part)

    def __truediv__(self, other):
        """
        Overloads the '/' operator.
        Formula: Multiply numerator and denominator by the complex conjugate.
        """
        if not isinstance(other, ComplexNumber):
            return NotImplemented
        
        denominator = other.real**2 + other.imag**2
        if denominator == 0:
            raise ZeroDivisionError("Cannot divide by a complex number equal to zero.")
            
        real_part = ((self.real * other.real) + (self.imag * other.imag)) / denominator
        imag_part = ((self.imag * other.real) - (self.real * other.imag)) / denominator
        return ComplexNumber(real_part, imag_part)
    
# Initialize two complex numbers: (3 + 4i) and (1 - 2i)
c1 = ComplexNumber(3, 4)
c2 = ComplexNumber(1, -2)

# Test __repr__ and __str__
print(f"c1 (str): {c1}")
print(f"c2 (str): {c2}")
print(f"Jupyter Repr evaluation: {repr(c1)}")

print("-" * 30)

# Test Math Operators
print(f"Addition (c1 + c2):       {c1 + c2}")
print(f"Subtraction (c1 - c2):    {c1 - c2}")
print(f"Multiplication (c1 * c2): {c1 * c2}")
print(f"Division (c1 / c2):       {c1 / c2}")
```
