### 🐍 Python `logging` Module: Core Theory

When building production-ready applications, `print()` statements quickly become messy and unmanageable. Python's built-in **`logging`** module provides a flexible, robust framework for tracking events, debugging errors, and auditing application behavior without cluttering the console.

---

#### 1. Why Use Logging Over `print()`?

* **Categorization:** Log messages are automatically tagged with severity levels (e.g., distinguishing a minor debug hint from a critical system crash).
* **Granular Control:** You can selectively turn log levels on or off globally without altering individual lines of code.
* **Multiple Destinations:** You can route logs to the console, flat files, email alerts, or external log servers simultaneously.
* **Rich Metadata:** Logs automatically include helpful context like timestamps, source file names, line numbers, and active threads.

---

#### 2. The 5 Core Logging Levels

By default, Python logs events at the **`WARNING`** level and above. Anything below `WARNING` (like `INFO` or `DEBUG`) will be silently ignored unless you explicitly reconfigure the root logger.

| Level | Numeric Value | When to Use It |
| :--- | :--- | :--- |
| **`DEBUG`** | `10` | Detailed diagnostic info, typically used only during development and troubleshooting. |
| **`INFO`** | `20` | Confirmation that things are working as expected (e.g., "Database connection established"). |
| **`WARNING`** | `30` | An indication that something unexpected happened, but the app is still running (e.g., "Low disk space"). |
| **`ERROR`** | `40` | A serious problem; the software was unable to perform a specific function or operation. |
| **`CRITICAL`** | `50` | A fatal error; the entire program may crash or stop running immediately. |

---

#### 3. The Architecture: How Logging Works

The `logging` module uses a modular architecture made up of four main components working together:

1. **Logger:** The entry point exposing the interface that your application code calls directly (e.g., `logger.info()`).
2. **Handler:** Directs the log records generated by the logger to their appropriate destination (e.g., `StreamHandler` for the console, `FileHandler` for text files).
3. **Formatter:** Specifies the layout, structure, and text styling of the final output string.
4. **Filter:** Provides fine-grained control over which log records get passed from logger to handler based on complex rules.

---

#### 4. Basic Configuration Example

To start logging immediately, use `logging.basicConfig()`. 

⚠️ **Jupyter Notebook Quick:** `basicConfig()` can only be called **once** per Python session. If you change its settings, you must restart your Jupyter kernel (`Kernel -> Restart`) for the new configurations to take effect.


```python
import logging

# configure basic logging setting
logging.basicConfig(level=logging.DEBUG)

# log message with different severity level
logging.debug("This is a debug message (saved to file).")
logging.info("Application successfully initialized.")
logging.warning("User attempted to access a deprecated API route.")
logging.error("This is an error message.")
logging.critical('This is a CRITICAL message')
```

**Configuration can be done once throughout a program so, to make another configuration `RESTART and only run the second cell`.**

```python
import logging

# Configure logging to write to a file with a specific format
logging.basicConfig(
    filename='app.log',
    filemode='w',
    level=logging.DEBUG, # Capture EVERYTHING from DEBUG level up
    format='%(asctime)s - %(levelname)s - %(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)

# log message with different severity level
logging.debug("This is a debug message (saved to file).")
logging.info("Application successfully initialized.")
logging.warning("User attempted to access a deprecated API route.")
logging.error("This is an error message.")
logging.critical('This is a CRITICAL message')
```
