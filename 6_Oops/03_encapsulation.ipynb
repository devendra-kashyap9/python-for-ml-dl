```python
# Encapsulation with Getter and Setter methods
# access specifier in other languages: private, protected, public

class Person:
    def __init__(self, name, age):
        self.name = name  # public var
        self.age = age  # public var, i.e. it can be accessed evn outside the class scope
        
person = Person('Eva', 27)
print(person.name)
print(person.age)
```

```python
print(dir(person))
```

```python
class Person:
    def __init__(self, name, age, sex):
        self._name = name  # protected var
        self.__age = age  # private var, i.e. it can't be accessed  outside the class scope
        self.sex = sex  # public var
        
person = Person('Eva', 27, 'Female')
```

```python
print(person.sex)
print(person._name)
print(person.__age) # private var not accessible outside the class
```
