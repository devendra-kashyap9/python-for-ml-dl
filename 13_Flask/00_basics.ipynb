### Flask

**Flask** is a lightweight, open-source Python web framework used to build web applications, REST APIs, and microservices. It is simple, flexible, and follows the **WSGI (Web Server Gateway Interface)** standard.

---

#### Features

- Lightweight and easy to learn
- Built-in development server
- URL routing
- Template engine (Jinja2)
- REST API development
- Supports extensions (SQLAlchemy, Flask-Login, etc.)

#### Common Components

| Component | Description |
|-----------|-------------|
| `Flask()` | Creates a Flask application |
| `@app.route()` | Maps a URL to a function |
| `request` | Access client request data |
| `render_template()` | Renders HTML templates |
| `jsonify()` | Returns JSON responses |
| `app.run()` | Starts the development server |

---

#### Advantages

- Simple and lightweight
- Easy to develop REST APIs
- Highly flexible
- Large community support
- Easy integration with databases and extensions

---

#### Disadvantages

- Not suitable for very large applications without proper architecture
- Fewer built-in features compared to Django
- Requires additional extensions for many functionalities

---

#### Applications

- Web applications
- REST APIs
- Machine Learning model deployment
- Microservices
- Backend development

---

#### Summary

- **Flask** is a lightweight Python web framework.
- It is mainly used for **web development** and **REST API development**.
- It is simple, flexible, and ideal for small to medium-sized applications.

### Flask - Important Components

#### 1. WSGI (Web Server Gateway Interface)

**WSGI (Web Server Gateway Interface)** is a standard interface between Python web applications (like Flask) and web servers (such as Gunicorn, uWSGI, Apache, or Nginx).

It allows Flask applications to communicate with web servers.

##### Flow

```
Client (Browser)
        │
        ▼
 Web Server (Nginx/Apache)
        │
     WSGI Server
        │
        ▼
 Flask Application
```

##### Benefits

- Standard communication interface
- Supports multiple web servers
- Improves scalability
- Enables deployment in production

---

#### 2. Jinja2 Template Engine

**Jinja2** is Flask's default template engine used to generate dynamic HTML pages by embedding Python-like expressions.

##### Example

```python
from flask import render_template

@app.route("/")
def home():
    return render_template("index.html", name="Devendra")
```

**index.html**

```html
<h1>Welcome {{ name }}</h1>
```

Output

```
Welcome Devendra
```

##### Features

- Variable substitution
- Loops
- Conditions
- Template inheritance
- Filters

---

#### 3. Routing

##### Definition

Routing maps a URL to a Python function.

```python
@app.route("/")
def home():
    return "Home Page"
```

---

#### 4. Request Object

##### Definition

The `request` object contains data sent by the client.

Example

```python
from flask import request

@app.route("/login", methods=["POST"])
def login():
    username = request.form["username"]
```

Used for

- Form data
- Query parameters
- Cookies
- Headers
- JSON data

---

#### 5. Response Object

Flask returns a response to the client.

```python
from flask import jsonify

@app.route("/api")
def api():
    return jsonify({"status": "success"})
```

---

#### 6. Templates

Templates are HTML files stored inside the **templates/** folder.

```
project/
│
├── app.py
├── templates/
│     └── index.html
```

Render using

```python
render_template("index.html")
```

---

#### 7. Static Files

Static files include

- CSS
- JavaScript
- Images

Stored inside

```
static/
    css/
    js/
    images/
```

Example

```html
<link rel="stylesheet" href="{{ url_for('static', filename='css/style.css') }}">
```

---

#### 8. URL Building

Flask uses `url_for()` to generate URLs dynamically.

```python
url_for("home")
```

Benefits

- Avoids hardcoding URLs
- Easy maintenance

---

#### 9. Sessions

Sessions store user data securely across requests.

```python
from flask import session

session["username"] = "Devendra"
```

---

#### 10. Cookies

Cookies store small amounts of data on the client's browser.

```python
response.set_cookie("user", "Devendra")
```

#### 11. Blueprints

Blueprints organize large Flask applications into modules.

Example

```
project/
│
├── app.py
├── auth/
├── admin/
├── blog/
```

Benefits

- Modular design
- Easy maintenance
- Reusable components
