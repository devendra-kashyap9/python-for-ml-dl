### Pandas DataFrame and Series

Series and DataFrame are the two foundational data structures of the Pandas library. A `Series` represents a single, one-dimensional column of data, while a `DataFrame` is a two-dimensional spreadsheet-like table composed of multiple grouped Series

```python
import pandas as pd
```

```python
## Series
# Series in one dimensional array like object that can hold any type of data.
data = [1,2,3,5,4]

ser = pd.Series(data)

print(ser)
print(type(ser))
```

```python
# Series from dictionary
data = {'a':1, 'b':2, 'c':3, 'd':4}
dict_series = pd.Series(data)
print(dict_series)
```

```python
data = [1,2,3,4]
index = ['a', 'b', 'c', 'd']

series = pd.Series(data=data,
                   index=index)
series
```

```python
## DataFrame
# dataframe from dictioanry of list

data={
    'Name': ['Alex', 'Bob', 'Chalres'],
    'Age': [24, 25, 24],
    'City': ['Paris', 'Geneva', 'Toronto']
}

df = pd.DataFrame(data)

print(df)
print(type(df))
```

```python
import numpy as np
np.array(df)  # converting DataFrame to an array
```

```python
# create DataFrame from list of dictionary
data = [
    {"Name": "Alex", "Age": 24, 'City': 'Paris'},
    {"Name": "Bob", "Age": 26, 'City': 'Geneva'},
    {"Name": "Charles", "Age": 24, 'City': 'Toronto'},
    {"Name": "Dia", "Age": 25, 'City': 'Auckland'}      
]

df = pd.DataFrame(data)

print(df)
print(type(df))
```

```python
# create DataFrame from list of dictionary
data = [
    {"Name": "Alex", "Age": 24, 'City': 'Paris'},
    {"Name": "Bob", "Age": 26, 'City': 'Geneva'},
    {"Name": "Charles", "Age": 24, 'City': 'Toronto'},
    {"Name": "Dia", "Age": 25, 'City': 'Auckland'}      
]

df = pd.DataFrame(data)

print(df)
print(type(df))
```

🔑 **Note**: Both loc and iloc can be used to acces the data, where loc is used to access the data with custom index, iloc is used for numerical/integer index.

**`loc` is inclusive of upper bound, while `iloc` is not**.
