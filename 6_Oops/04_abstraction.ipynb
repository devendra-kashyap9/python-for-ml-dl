### Abstraction in Python: Theory & Core Concepts

**Abstraction** is a core pillar of Object-Oriented Programming (OOP). It is the process of hiding internal complexity and showing only the essential features of an object to the user. 

Think of a car: you only need to interact with the steering wheel, gas pedal, and brakes to drive it. You do not need to understand the thermodynamics of the internal combustion engine or how the electronic control unit processes fuel injection. The interface is **abstracted** away from the execution details.



---

#### 1. Why Do We Need Abstraction?

1. **Reduces Complexity:** Users interact with a clean, simple interface without being overwhelmed by underlying code logic.
2. **Enforces a Framework/Contract:** It acts as a blueprint, forcing subclasses to implement specific methods so that your software architecture stays consistent.
3. **Improves Maintainability:** You can change the complex internal workings of a system without changing how the user interacts with it.

---

#### 2. Abstraction vs. Encapsulation

While they sound similar and work hand-in-hand, they solve completely different problems:

* **Encapsulation:** Focuses on **Data Hiding**. It bundles data and restricts direct access to it (using private variables like `__variable`). It answers the question: *How do I protect my data?*
* **Abstraction:** Focuses on **Hiding Complexity**. It focuses on the outer interface rather than internal implementation. It answers the question: *What does this object do, rather than how does it do it?*

---

#### 3. Implementing Abstraction using the `abc` Module

Python does not support abstract classes by default. To implement abstraction, you must import the **`abc` (Abstract Base Classes)** module and use the `@abstractmethod` decorator.

An **Abstract Class** is a class that contains one or more abstract methods. **Crucial Rule:** You can never create an object (instantiate an instance) directly from an abstract base class.

```python
from abc import ABC, abstractmethod

# 1. Define the Abstract Base Class (The Blueprint)
class Vehicle(ABC):
    
    @abstractmethod
    def start_engine(self):
        """Abstract method; no implementation details here."""
        pass

    @abstractmethod
    def turn_off(self):
        pass

````python
from abc import ABC, abstractmethod

# abstract class
class Vehicle(ABC):
    def drive(self):
        print('The vehicle is used for driving.')
        
    @abstractmethod
    def start_engine(self):
        pass
    
# child class
class Car(Vehicle):
    def start_engine(self):
        print('Car Engine started')
        
def operate_vehicle(vehicle):
    vehicle.start_engine()
    vehicle.drive()
    
# instantiation
car = Car()
operate_vehicle(car)
````
